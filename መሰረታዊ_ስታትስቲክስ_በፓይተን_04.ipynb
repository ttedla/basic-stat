# የሳጥን ገበታ (ቦክስፕሎት) 

የሳጥን ገበታ ወይም ቦክስፕሎት የምንለው ስለውሂባችን አምስቱን የቁጥር ማጠቃለያዎች (ፋይቭ ነምበርስ ሰመሪ) የሚባሉትን በገበታ መልክ ያሳየናል። አምስቱ የቁጥር ማጠቃለያዎች የሚባሉት፡
* አጋማሽ - 50% የሚሆነው የውሂባችን እሴት ከአጋማሹ በታች መሆኑን የሚነግረን። 
* ትንሹን እሴት - በውሂባችን የሚገኘውን ትንሹን እሴት 
* ትልቁን እሴት  - በውሂባችን የሚገኘውን ትልቁን እሴት 
* ኳንታይል ፩ (Q1) - የታችኛው 50% ውሂብ አጋማሽ ነው። (25% የሚሆነው እሴት ከ ኳ፩ በታች ነው) 
* ኳንታይል ፫ (Q3) - የላይኛው 50% ውሂብ አጋማሽ ነው። (25% የሚሆነው እሴት ከ ኳ፫ በላይ ነው)

*** 
ኢንተር ኳንታይ ሬንጅ (IQR) የምንለው የ ኳ፫ ሲቀነስ ኳ፩ የሚገኘው እሴት ነው። ይህ ማለት 50% የሚሆነው ውሂባችን በ ኳ፩ እና በኳ፫ መሀከል ይገኛል ማለት ነው። 
*** 
የሳጥን ገበታ ከዚህ በተጫማሪ ስለውሂባችን ስብጥር ዘመምነት(ስኪውድነስ) መረጃ ይሰጠናል። የሚከተለው የሳጥን ገበታ ምሳሌ ነው። 

![የሳጥን ገበታ](./ውሂብ/boxplot.jpg "የሳጥን ገበታ ምሳሌ")

ከ ሳጥን ገበታው መመልከት እንደምንችለው 50% ውሂባችን የሚገኘው በ ሩብ አንድ እና በሩብ ሶስት መካከል ነው። የበትሩ ርዝማኔ ከአጋማሹ ጋር በመጣመር ስብጥሩ ላይ ዘመም (ፓዘቲቭ ስኪውድ) ወይም ታች ዘመም (ኔጋቲቭ ስኪውድ) መሆኑን ይጠቁማል።  

# የሳጥን ገበታ በምሳሌ

በተለያዩ የኢትዮጵያ ከተሞች አማካይ የነዋሪዎችን የቀን ገቢ ሰበሰብን እንበል (26 ከተሞች) እና ውሂቡ የሚከተለው ነው። 52, 55, 56, 64, 66, 66, 67, 69, 71, 72, 72, 73, 73, 74, 74, 74, 74, 71, 72, 72, 73, 73, 74, 74, 74, 74 ነው እንበል።  ስለውሂባችን ግንዛቤ ለማግኘት የሳጥን ገበታ በመሳል መመልከት እንችላለን። 

```python
# አስፈላጊ ጥቅሎችን እንጥራ
from matplotlib.font_manager import FontProperties
import matplotlib.pyplot as plt
ውሂብ_ነቁጦች = (69, 75, 76, 77, \
            71, 72, 72, 73, 73, 74, 74, 74, 74, 
            71, 72, 72, 73, 73, 74, 74, 74, 74)
ቅርጸ_ቁምፊ = FontProperties('Nyala')

%matplotlib inline 
መሳያ = plt.figure() 
bp = plt.boxplot(ውሂብ_ነቁጦች)
plt.title('አማካኝ የቀን ገቢ', fontproperties=ቅርጸ_ቁምፊ, fontsize=20)
plt.tick_params(
    axis='x',          # ለውጡ x-axisን ብቻ እንዲመለከተው ማድረግ 
    which='both',      # ዋና እና ንዑስ ጭረቶችን እንዲመለከት ማድረግ 
    bottom=False,      # የታችኛውን ጭረት ማጥፋት 
    top=False,         # የላይኛውን (ካለ) ጭረት ማጥፋት 
    labelbottom=False) # ከታች የሚገኙ ሌብሎችን ማጥፋት
plt.ylabel('የቀን ገቢ (በብር)', fontproperties=ቅርጸ_ቁምፊ, fontsize=16)
መሳያ.savefig("./ውሂብ/የሳጥን_ገበታ_01.pdf", bbox_inches='tight')
```

ከ ሳጥን ገበታው እንደምንመለከተው የውሂብ ስብጥራችን የተማከለው በ 73.5 ላይ ነው። ይህም የተመለከተው በ አጋማሹ ነው። ስብጥሩ ደግሞ (በ ኢንተር ኳርታይል ሬንጅ እንደተመለከተው) 2 ነው። አካላይነቱ ወይም ሬንጁ ደግሞ 8 ነው። ትንሹ እሴት 69 ሲሆን ትልቁ የውሂባችን እሴት ደግሞ 77 ነው። 50% የሚሆነው የውሂባችን እሴት የሚገኘው በ 72 እና 74 መካከል ነው። የሳጥን ገበታን በመጠቀም ለትንተና ጠቃሚ የሆኑ መረጃዎችን በዚህ መልክ ማግኘት ይቻላል።   

# ፋይልን በማንበብ የሳጥን ገበታን መስራት 

```python
# የአማርኛ ቅርጸ ቁምፊ ለመጠቀም የሚረዳ ጥቅል ነው 
from matplotlib.font_manager import FontProperties
# አሞሌ ገበታውን ለመሳል የምንጠቀምበት ጥቅል ነው። 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# ውሂቡን ለማንበብ የምንጠቀምበት ጥቅል ነው። 
import pandas as pd
#import matplotlib.ticker as ticker
# ኒያላ የሚለውን ለቅርጸ ቁምፊም ንብረት እንድርግ። 
ቅርጸ_ቁምፊ = FontProperties('Nyala')
# ኒያላ የሚለውን ቅረጸቁሚፍ እንዲጠቀም ለማሳሰብ 
sns.set(font="Nyala")
#tick_spacing = 200
# ገበታውን እዚሁ ላይ እንዲያሳየን  
%matplotlib inline 
# ውሂቡን ውሂብ በሚለው ተለዋዋጭ ላይ እናንብብ
ውሂብ = pd.read_csv('./ውሂብ/የቤት_ኪራይ_በካሬ.csv')

መሳያ = plt.figure() 
# ሲቦርንን በመጠቀም ገበታውን እንሳል 
bplot = sns.boxplot(y='ዋጋ', x='ሰፈር', 
                 data=ውሂብ, 
                 width=0.5,
                 palette="colorblind")
# የገበታውን ርዕስ እንደምንፈልገው እናስተካክል 
bplot.axes.set_title("የንግድ ቤት ኪራይ ዋጋ በካሬ",fontproperties=ቅርጸ_ቁምፊ,
                   fontsize=16)
#bplot.yaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
# የተሳለውን ገበታ በፋል እናስቀምጠው
መሳያ.savefig("./ውሂብ/የሳጥን_ገበታ_02.pdf", bbox_inches='tight')
```

ከገበታው እንደምንመለከተው የመገናኛ ውሂብ ላይ ዘመም (ፓዘቲቭሊ ስኪውድ) እንዲሁም የሳሪስ ደግሞ ታች ዘመም (ኔጋቲቭሊ ስኪውድ) መሆኑን በትሩን እና አጋማሹን በመመልከተ መናገር ይቻላል። የማርካቶ የቤት ኪራይ በካሬ አጋማሹ 3100 ሲሆን አቃቂ አካባቢ አጋማሹ እስከ 750 ብር በካሬ ነው። ተጨማሪ ንፅፅራዊ ትንተናዎችን ገበታውን በመመልከት ማካሄድ ይቻላል። መርካቶ አካባቢ ቤት በካሬ ውድ ቢሆንም ከፍተኛ የሆነ የዋጋ ልዩነት በአንጻራዊነት እንደሚታይበት መገመት ይቻላል። በቁጥር ለማረጋገጥ ኢንተርኳርታይል ሬንጁን እና ሬንጁን መቀመር ነው።  

```python
# ገበታውን በሰፈር ሳይሆን በወር ሰብስቦ መሳል ይቻላል። 
# የአማርኛ ቅርጸ ቁምፊ ለመጠቀም የሚረዳ ጥቅል ነው 
from matplotlib.font_manager import FontProperties
# አሞሌ ገበታውን ለመሳል የምንጠቀምበት ጥቅል ነው። 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# ውሂቡን ለማንበብ የምንጠቀምበት ጥቅል ነው። 
import pandas as pd
#import matplotlib.ticker as ticker
# ኒያላ የሚለውን ለቅርጸ ቁምፊም ንብረት እንድርግ። 
ቅርጸ_ቁምፊ = FontProperties('Nyala')
# ኒያላ የሚለውን ቅረጸቁሚፍ እንዲጠቀም ለማሳሰብ 
sns.set(font="Nyala")
#tick_spacing = 200
# ገበታውን እዚሁ ላይ እንዲያሳየን  
%matplotlib inline 
# ውሂቡን ውሂብ በሚለው ተለዋዋጭ ላይ እናንብብ
ውሂብ = pd.read_csv('./ውሂብ/የቤት_ኪራይ_በካሬ.csv')

መሳያ = plt.figure() 
# ሲቦርንን በመጠቀም ገበታውን እንሳል 
bplot = sns.boxplot(y='ዋጋ', x='ወር', 
                 data=ውሂብ, 
                 width=0.5,
                 palette="colorblind")
# የገበታውን ርዕስ እንደምንፈልገው እናስተካክል 
bplot.axes.set_title("የንግድ ቤት ኪራይ ዋጋ በካሬ",fontproperties=ቅርጸ_ቁምፊ,
                   fontsize=16)
# በ ረድፍ ያሉትን የወር ስሞች እንዲታዩ ለማድረግ 
plt.xticks(rotation=90)
# የተሳለውን ገበታ በፋል እናስቀምጠው

መሳያ.savefig("./ውሂብ/የሳጥን_ገበታ_03.pdf", bbox_inches='tight')
```

የቤት ኪራይ በአማካኝ በየትኛው ወር እንደሚወደድ ወይም እንደሚረክስ ከገበታው መመልከት ይቻላል። በተጨማሪም በ አልማዝ ቅርጽ የተመለከቱት ጽንፍ (አውትላየር) ይባላሉ። ጽንፍ ብዙ ግዜ 1.5IRQ በመጠቀም የሚሰላ ነው። ለጊዜው እዚህ ላይ ይብቃ።

* የውሂቡን የስብጥር ዘመምነት ማወቅ ይቻላል። 
* አጋማሹን ማወቅ ይቻላል። 
* ትንሹ እና ትልቁን እሴት ማወቅ ይቻላል። 
* ኢንተር ኳርታይ ሬንጁን ማወቅ ይቻላል። ( የ $3^{ተኛ}$ ሩቡን እና የ $1^{ኛ}$ ሩቡን በመጠቀም) 
* ሬንጁን ወይም አካላይነቱን ማወቅ ይቻላል።
